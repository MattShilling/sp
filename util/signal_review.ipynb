```python
import pandas as pd
import matplotlib.pyplot as plt

def init_plt():    
    plt.rcParams.update(plt.rcParamsDefault)
    plt.rcParams.update({
        'legend.fontsize': 5,
        'legend.handlelength': 1,
    })
    %matplotlib notebook
    plt.rcParams['figure.dpi'] = 150

def grid(plt):
    plt.grid(which='major', linestyle='--',)
    plt.grid(which='minor', linestyle='--', fillstyle='full', alpha=0.2)
    plt.minorticks_on()
    
init_plt()
```

```python
df = pd.read_csv('../samples.csv')
```

```python
plt.figure()

plt.plot(df['time'],
         df['amplitude'],
         marker='.')

grid(plt)
```
