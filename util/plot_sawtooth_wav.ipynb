```python
import librosa
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the audio file
file_path = '../build/src/tests/sawtooth_stereo.wav'

# y: audio time series, sr: sampling rate
y, sr = librosa.load(file_path)

# Compute the short-time Fourier transform (STFT).
D = librosa.stft(y)

# Convert the STFT to magnitude (amplitude) spectrogram.
mag_D = np.abs(D)

# Convert the magnitude spectrogram to Decibels.
db_mag_D = librosa.amplitude_to_db(mag_D, ref=np.max)
```

```python
# Display the spectrogram
plt.figure()
plt.figure(figsize=(12, 6))
D_highres = librosa.stft(y, hop_length=256, n_fft=4096)
S_db_hr = librosa.amplitude_to_db(np.abs(D_highres), ref=np.max)
librosa.display.specshow(S_db_hr, sr=sr, hop_length=256, x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('Spectrogram')
```
